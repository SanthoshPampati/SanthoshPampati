```python
a = folium.map.FeatureGroup()
for long,lat, in zip (df_200['X'],df_200['Y']):
    a.add_child(folium.features.CircleMarker([lat,long],radius = 5,
                                             color = "green", fill = True, 
                                             fill_color = 'red',fill_opacity = 0.4))
sanfran.add_child(a)
```

```python
sanfran_map = folium.Map(location=[37.77,-122.42], zoom_start=12)
for long,lat, in zip (df_200['X'],df_200['Y']):
    folium.features.CircleMarker([lat, long], radius = 6, color = 'blue', 
                            fill = True, fill_color = 'red',
                            fill_opacity = 0.6).add_to(sanfran_map)

sanfran_map # other simple way
```

```python
a = folium.map.FeatureGroup()
for long,lat in zip (df_200['X'],df_200['Y']):
    a.add_child(folium.features.CircleMarker([lat,long],radius = 5,
                                             color = "green", fill = True, 
                                             fill_color = 'red',fill_opacity = 0.4))

for long, lat, labels in zip (df_200['X'], df_200['Y'], df_200['Category']):
    folium.Marker([lat, long], popup = labels).add_to(sanfran)
sanfran.add_child(a)
```

```python
from folium import plugins 
sanfran_map = folium.Map(location = [37.77,-122.42], zoom_start = 12)
b = plugins.MarkerCluster().add_to(sanfran_map)


for long, lat, labels in zip (df_200['X'], df_200['Y'], df_200['Category']):
    folium.Marker([lat, long], popup = labels).add_to(b)

sanfran_map
```

```python
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                     sheet_name='Canada by Citizenship',
                     skiprows=range(20),
                     skipfooter=2)
```

```python
df_can.head()
```

```python
#df_can.drop(['Type','Coverage','AREA','REG','DEV','DevName'], axis = 1, inplace = True)
df_can.rename(columns = {'OdName':'Countrys','AreaName':'Continent','RegName':'Region'}, inplace = True)
df_can['Total'] = df_can.sum(axis = 1)
df_can.head()
```

```python
df_can.drop(['Type','Coverage','AREA','REG','DEV','DevName'], axis = 1, inplace = True)
```

```python
df_can.columns = list(map(str, df_can.columns))
df_can.head()
```

```python
!wget --quiet https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/world_countries.json -O world_countries.json
    
print('GeoJSON file downloaded!')
```

```python
world_geo = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/world_countries.json'
world_map = folium.Map(location =[0,0], zoom_start = 2, tiles = 'Mapbox Bright')
world_map
```

```python
#world_map = folium.Map(tiles = 'Mapbox Bright')
world_map.choropleth(geo_data = world_geo, data = df_can, 
                     columns = ['Countrys', 'Total'], key_on = 'feature.properties.name',
                     fill_color='YlOrRd', fill_opacity=0.7, 
                     line_opacity=0.2, legend_name='Immigration to Canada')
world_map
```

```python
t = np.linspace(df_can['Total'].min(),df_can['Total'].max(), 6, dtype = int)
```

```python
t = t.tolist()
t[-1] = t[-1]+1
t
```

```python
world_map = folium.Map(location =[0,0], zoom_start = 2, tiles = 'Mapbox Bright')
world_map.choropleth(geo_data = world_geo, data = df_can, 
                     columns = ['Countrys', 'Total'], key_on = 'feature.properties.name',
                     fill_color='YlOrRd', threshold_scale = t, fill_opacity=0.7, 
                     line_opacity=0.2, legend_name='Immigration to Canada', reset = True)
world_map
```
